```python
import requests
```

```python
def request_camara_deputados(endpoint, params=None):
    base_url = "https://dadosabertos.camara.leg.br/api/v2/"
    url = f"{base_url}{endpoint}"
    response = requests.get(url, params=params)
    
    status_code = response.status_code
    if status_code != 200:
        raise Exception(f"API request failed with status code {status_code}\nResponse: {response.text}")

    return response.json()
```

```python
request_camara_deputados("deputados", params={"siglaUf": "SP", "ordem": "ASC", "ordenarPor": "nome"})
```
