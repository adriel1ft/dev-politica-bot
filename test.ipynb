```python
import requests

url = "https://dadosabertos.camara.leg.br/api/v2/proposicoes"
params = {
    "ano": "2025",
    "siglaTipo": "PEC"
    }
headers = {
    "accept": "application/json"
}

response = requests.get(url, params=params, headers=headers)
```

```python
import pandas as pd
dados = response.json()["dados"]
df = pd.DataFrame(dados)
df.to_csv("proposicoes.csv", index=False)
df.head()  
```

```python
# filepath: c:\Users\biabc\OneDrive\Documents\hackaton\dev-politica-bot\test.ipynb
from langchain_openai import ChatOpenAI
from langchain_experimental.agents import create_pandas_dataframe_agent
from dotenv import load_dotenv
import os

load_dotenv()  # Carregar variáveis de ambiente do arquivo .env

API_KEY = os.getenv("OPENAI_API_KEY")
llm = ChatOpenAI(openai_api_key=API_KEY)

# Criar o agente que entende linguagem natural e consulta o DataFrame
agent = create_pandas_dataframe_agent(llm, df, verbose=False)

# Exemplo de pergunta
pergunta = "O que diz a ementa da PEC 8/2025?"
resposta = agent.run(pergunta)
print(resposta)
```

```python
from gtts import gTTS

# Converter a resposta em áudio
tts = gTTS(resposta, lang='pt')
tts.save("resposta.mp3")
print("Áudio salvo como resposta.mp3")
```
